```python
# Change column names (Note: Replacing multiple characters in column resulted in error. Not sure how to fix.)
df_poc.columns = df_poc.columns.str.replace('Budget % - ','B_')
df_poc

```

```python
df_poc.dtypes
```

```python
# Replace ranges as an average of the range & converted NaN to Zero
df_poc['B_Editorial'] = df_poc['B_Editorial'].replace("0-10", "5").replace("11-20", "15").replace("21-30", "25").replace("31-40", "35").replace("41-50", "45").replace("51-60", "55").replace("61-70", "65").replace("71-80", "75").replace("81-90", "85").replace("91-100", "95").fillna(0)

df_poc['B_Revenue Generation'] = df_poc['B_Revenue Generation'].replace("0-10", "5").replace("11-20", "15").replace("21-30", "25").replace("31-40", "35").replace("41-50", "45").replace("51-60", "55").replace("61-70", "65").replace("71-80", "75").replace("81-90", "85").replace("91-100", "95").fillna(0)

df_poc['B_Product/Technology'] = df_poc['B_Product/Technology'].replace("0-10", "5").replace("11-20", "15").replace("21-30", "25").replace("31-40", "35").replace("41-50", "45").replace("51-60", "55").replace("61-70", "65").replace("71-80", "75").replace("81-90", "85").replace("91-100", "95").fillna(0)

df_poc['B_Administration'] = df_poc['B_Administration'].replace("0-10", "5").replace("11-20", "15").replace("21-30", "25").replace("31-40", "35").replace("41-50", "45").replace("51-60", "55").replace("61-70", "65").replace("71-80", "75").replace("81-90", "85").replace("91-100", "95").fillna(0)
df_poc['B_Administration']


```

```python
# Convert Budget % - Editorial/ Revenue Generation / Product/Technology / Administration to integer

df_poc[['B_Editorial', 'B_Revenue Generation', 'B_Product/Technology', 'B_Administration']] = df_poc[['B_Editorial', 'B_Revenue Generation', 'B_Product/Technology', 'B_Administration']].astype(int)

df_poc.dtypes
```

```python
# Calculate mean for each of the four sections
df_budget = df_poc[['B_Editorial', 'B_Revenue Generation', 'B_Product/Technology', 'B_Administration']].reset_index()

#replace 0 with NaN for mean calculation
df_budget = df_budget.replace(0, np.NaN)

# Create new dataframe with mean of budgets
df_budget_mean = df_budget[['B_Editorial', 'B_Revenue Generation', 'B_Product/Technology', 'B_Administration']].mean().reset_index()

df_budget_mean
```

```python
#change column names
df_budget_mean = df_budget_mean.rename(columns={"index": "Category", 0: "Mean_Percentage"})
df_budget_mean
```

```python
df_budget_mean['Mean_Percentage'] = df_budget_mean['Mean_Percentage'].round(1)
df_budget_mean

#linear bar graph for visual?
```

```python
#export budget as csv
df_budget_mean.to_csv('mean_budget.csv', index=False)
```
